```python
# import libraries
import os 
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fastai.vision.all import *
```

```python
# Each of these directories contain examples of different class
len(os.listdir("..\dataset")), os.listdir("..\dataset")
```

```python
# Lets see how many example we have for each class
d = {}
for c in os.listdir("..\dataset"):
    num = len(os.listdir("..\\dataset\\"+c))
    print(c,':', num)
    d[c]=num
```

```python
# make a barplot
names = list(d.keys())
values = list(d.values())


fig, ax = plt.subplots(figsize =(16, 50))
 
# Horizontal Bar Plot
ax.barh(names, values)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')

 
# Add Plot Title
ax.set_title('Classnames and number of trainin',
             loc ='left', )
```
