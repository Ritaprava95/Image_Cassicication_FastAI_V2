```python
# import libraries
import os 
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fastai.vision.all import *
```

```python
# Each of these directories contain examples of different class
print(len(os.listdir("..\dataset")), os.listdir("..\dataset"))
```

```python
# Lets see how many example we have for each class
d = {}
for c in os.listdir("..\dataset"):
    num = len(os.listdir("..\\dataset\\"+c))
    print(c,':', num)
    d[c]=num
```

```python
# make a barplot
names = list(d.keys())
values = list(d.values())


fig, ax = plt.subplots(figsize =(16, 50))
 
# Horizontal Bar Plot
ax.barh(names, values)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')


# Add Plot Title
ax.set_title('Classnames and number of trainin',
             loc ='left', )
```

```python
base_path = "E:\\Work\Data_Science\\Projects\\Image_Classification_FastAI_V2\\dataset"
data = pd.DataFrame(columns=['image_name', 'label'])

image_names = []
labels = []
for c in os.listdir("..\dataset"):
    for i in os.listdir("..\\dataset\\"+c):
        image_names.append(c+"\\"+i)
        labels.append(c)
data['image_name'] = image_names
data['label'] = labels
```

```python
data.head()
```

```python
# Split train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['image_name'], data['label'], test_size=0.2, stratify=data['label'], random_state=12345)
```

```python
X_train[:10]
```

```python
y_train[:10]
```

```python
# Make and save training data
train_data = pd.DataFrame()
train_data['image_name'] = X_train
train_data['label'] = y_train
train_data.to_csv(r"..\train_data.csv", index=False)
```

```python
# Make and save test data
test_data = pd.DataFrame()
test_data['image_name'] = X_test
test_data['label'] = y_test
test_data.to_csv(r"..\test_data.csv", index=False)
```

```python
train_data = pd.read_csv(r"..\train_data.csv")
train_data.head()
```

## Dataloader for multiclass classification

```python
# Prepare fastai training data

fastai_vision_classifier_data = (
    ImageDataLoaders.from_df(
        path=base_path,
        df=train_data,
        valid_pct=0.3,
        random_seed=123456,
        fn_col="image_name",
        label_col="label",
        item_tfms=Resize(200, method=ResizeMethod.Pad, pad_mode=PadMode.Zeros),
        # ds_tfms=RandTransform() #for data augmentation
    )
)
```
