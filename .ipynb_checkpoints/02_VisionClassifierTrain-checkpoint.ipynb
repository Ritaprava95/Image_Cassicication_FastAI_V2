```python
# import packages
import sys, os, joblib
import numpy as np
import pandas as pd
from fastai.vision.all import *
```

```python
import torch
torch.cuda.get_device_name()
```

```python
fastai_vision_classifier_data = torch.load(r"..\fastai_data\fastai_vision_classifier_data.pkl")
```

```python
fastai_vision_classifier_data.show_batch()
```

```python
# make batch size = 16
fastai_vision_classifier_data.bs = 16
```

```python
# available background models
print(dir(models))
```

```python
# make the model
learner = vision_learner(
    fastai_vision_classifier_data,
    models.resnet18,
    metrics=accuracy
)
```

```python
# find best learning rate when only top few layers are unfreezed
learner.lr_find()
```

```python
# train the model for only last few layers
learner.fit_one_cycle(5, 0.002)
```

```python
# unfreeze all the model weights
learner.unfreeze()
learner.lr_find()
```

```python
# train the model for all the layers
learner.fit_one_cycle(5, slice(0.0001, 0.01))
```

```python
# save the model
learner.export(r"..\model\vision_classifier_model")
```
